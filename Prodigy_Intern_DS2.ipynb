STEP 1 : LOAD THE DATA

```python

import pandas as pd

# Load the Titanic dataset
titanic_url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(titanic_url)
df.head()
```

STEP 2 : DATA CLEANING

```python
# STEP 2 : DATA CLEANING

# Check for missing values
df.isnull().sum()
```

```python
# STEP 3:HANDLING MISSING VALUES

# Fill missing values
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)

# Create a new feature for Cabin
df['Has_Cabin'] = df['Cabin'].apply(lambda x: 0 if pd.isnull(x) else 1)

# Drop the Cabin column
df.drop('Cabin', axis=1, inplace=True)
```

```python
#STEP 5:CONVERT CATEGORICAL VARIABLE TO NUMERICAL

# Convert 'Sex' to numerical
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})

# Convert 'Embarked' to numerical
df['Embarked'] = df['Embarked'].map({'S': 0, 'C': 1, 'Q': 2})
```

```python
# Drop non-numerical columns
df_numeric = df.drop(['Name', 'Ticket'], axis=1)
```

STEP 3 : EXPLORATORY DATA ANALYSIS(EDA)

VISUALIZATION

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set a different style
sns.set(style='darkgrid')

# Plot survival rate by sex with new colors and font
plt.figure(figsize=(10, 6))
sns.barplot(x='Sex', y='Survived', data=df, palette='viridis')
plt.title('Survival Rate by Sex', fontsize=16, fontweight='bold', color='darkblue')
plt.xlabel('Sex', fontsize=14, fontweight='bold', color='darkblue')
plt.ylabel('Survival Rate', fontsize=14, fontweight='bold', color='darkblue')
plt.xticks(fontsize=12, color='darkblue')
plt.yticks(fontsize=12, color='darkblue')
plt.show()

# Plot survival rate by Pclass with new colors and font
plt.figure(figsize=(10, 6))
sns.barplot(x='Pclass', y='Survived', data=df, palette='magma')
plt.title('Survival Rate by Pclass', fontsize=16, fontweight='bold', color='darkgreen')
plt.xlabel('Passenger Class', fontsize=14, fontweight='bold', color='darkgreen')
plt.ylabel('Survival Rate', fontsize=14, fontweight='bold', color='darkgreen')
plt.xticks(fontsize=12, color='darkgreen')
plt.yticks(fontsize=12, color='darkgreen')
plt.show()

# Plot age distribution by survival with new colors and font
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Age', hue='Survived', multiple='stack', palette='coolwarm')
plt.title('Age Distribution by Survival', fontsize=16, fontweight='bold', color='purple')
plt.xlabel('Age', fontsize=14, fontweight='bold', color='purple')
plt.ylabel('Count', fontsize=14, fontweight='bold', color='purple')
plt.xticks(fontsize=12, color='purple')
plt.yticks(fontsize=12, color='purple')
plt.legend(title='Survived', fontsize=12, title_fontsize=14)
plt.show()

# Plot fare distribution by survival with new colors and font
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Fare', hue='Survived', multiple='stack', palette='cubehelix')
plt.title('Fare Distribution by Survival', fontsize=16, fontweight='bold', color='maroon')
plt.xlabel('Fare', fontsize=14, fontweight='bold', color='maroon')
plt.ylabel('Count', fontsize=14, fontweight='bold', color='maroon')
plt.xticks(fontsize=12, color='maroon')
plt.yticks(fontsize=12, color='maroon')
plt.legend(title='Survived', fontsize=12, title_fontsize=14)
plt.show()
```

STEP 4 : CORRELATION MATRIX

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Select only numerical columns before calculating correlation
numerical_df = df.select_dtypes(include=['number'])

# Set the style and color palette
sns.set(style='white', palette='bright')

# Correlation matrix with enhanced visuals
plt.figure(figsize=(12, 8))
sns.heatmap(numerical_df.corr(), annot=True, cmap='Spectral', fmt='.2f', annot_kws={"size": 12, "weight": "bold", "color": "black"})
plt.title('Correlation Matrix', fontsize=18, weight='bold', color='navy')
plt.xticks(fontsize=12, weight='bold', color='navy')
plt.yticks(fontsize=12, weight='bold', color='navy')
plt.show()
```
